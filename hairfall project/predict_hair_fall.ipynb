```python
cat_col = df.select_dtypes(include = 'O')
num_col = df.select_dtypes(exclude = "O")
```

```python
cat_col
```

```python
from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
encoder.fit(cat_col)
```

```python
transform_df = encoder.transform(cat_col)
```

```python
encoder.get_feature_names_out()
```

```python
cat_col = pd.DataFrame(transform_df , columns = encoder.get_feature_names_out())
```

```python
transform_df.shape
```

```python
final_df = pd.concat([ num_col,cat_col ] , axis = 1)
```

```python
final_df
```

```python
final_df.shape
```

```python
final_df.dtypes
```

```python
x = final_df.drop("Hair Loss" , axis = 1)
```

```python
y = final_df[['Hair Loss']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.svm import SVC
```

```python
lgr = LogisticRegression(C=1.0)
lr = LinearRegression()
```

```python
# lgr.fit(x_train,y_train)
lr.fit(x_train,y_train)
```

```python
print(lr.score(x_test,y_test))
```

```python
model = DecisionTreeClassifier(criterion = "gini" , max_depth = 3 , random_state = 42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
svmmodel = SVC(kernel = "linear" , C = 1).fit(x_train,y_train)
```

```python
print(svmmodel.score(x_test,y_test))
```
